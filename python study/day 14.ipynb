```python
class sample:
    
    def __init__(self,n):
        self.n = n
    def square(self):
        for i in range(self.n):
            time.sleep(0.2)
            print(threading.current_thread())
    #         print(f'square : {i*i}')
    def cube(self):
        for i in range(self.n):
            time.sleep(0.5)
    #       print(f'cube : {i*i*i}')
            print(threading.current_thread())

    
    
```

```python
o = sample(10)
t1 = threading.Thread(target=o.square)
t2 = threading.Thread(target=o.cube)

```

```python
t2.start()
t1.start()
t1.join()
```

```python
class ticketreserve:
    total = 0
    
#     def __init__(self,total):
#         self.total = total
        
    def resreve(self,number):
#         print("%s reached at %f" % (threading.current_t.name,time.time()))
        if ticketreserve.total >= number:
            ticketreserve.total -= number
            print(f"{threading.current_thread()} is reserved {number} tickets")
            
        else:
            print(f"{number} tickets not for {threading.current_thread()} ")
```

```python
print(f"{threading.current_thread().name} started")
ticketreserve.total = 10

tom = ticketreserve()
gery = ticketreserve()
dora = ticketreserve()
buji = ticketreserve()

tomT = threading.Thread(target=tom.resreve,args=(5,),name='tom')
geryT = threading.Thread(target=gery.resreve,args=(5,),name='gery')
doraT = threading.Thread(target=dora.resreve,args=(5,),name='dora')
bujiT = threading.Thread(target=buji.resreve,args=(5,),name='buji')

tomT.start()
geryT.start()
doraT.start()
bujiT.start()
time.sleep(3)
print("balance ticket %d" % (ticketreserve.total))
print(f"{threading.current_thread().name} completed")

                         
                         
```
